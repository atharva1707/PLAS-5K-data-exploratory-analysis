```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import concurrent.futures
from google.colab import files
```

```python
df = pd.read_csv('5000_final.csv')
```

```python
# Extract columns of interest
binding_affinity = df["binding_affinity (kcal/mol)"]
binding_affinity_sd = df["binding_affinity_sd (kcal/mol)"]
electrostatic = df["electrostatic (kcal/mol)"]
electrostatic_sd = df["electrostatic_sd (kcal/mol)"]

# slice the data
binding_affinity = binding_affinity[0:100]
binding_affinity_sd = binding_affinity_sd[0:100]
electrostatic = electrostatic[0:100]
electrostatic_sd = electrostatic_sd[0:100]
# Create error bar plots for binding affinity and electrostatic energy
plt.figure(figsize=(10, 6))

# Plot for binding affinity
plt.errorbar(
    range(len(binding_affinity)),
    binding_affinity,
    yerr=binding_affinity_sd,
    label="Binding Affinity",
    marker='o',
    linestyle='-',
    capsize=5
)


plt.title("Error Analysis: Binding Affinity ")
plt.xlabel("Data Points")
plt.ylabel("Energy (kcal/mol)")
plt.legend()
plt.grid(True)

# Show the plot (or save it to a file using plt.savefig)
plt.show()

# Plot for electrostatic energy
plt.errorbar(
    range(len(electrostatic)),
    electrostatic,
    yerr=electrostatic_sd,
    label="Electrostatic Energy",
    marker='s',
    linestyle='--',
    capsize=5
)

plt.title("Error Analysis: Electrostatic Energy")
plt.xlabel("Data Points")
plt.ylabel("Energy (kcal/mol)")
plt.legend()
plt.grid(True)
```
